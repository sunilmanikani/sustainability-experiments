```python
from google.colab import userdata
DEEP_SEEK_API_KEY = userdata.get('DEEP_SEEK_API_KEY')
from openai import OpenAI
client = OpenAI(api_key=DEEP_SEEK_API_KEY, base_url="https://api.deepseek.com")
prompt = """
		    Gangapur
         /   \
    (25)/     \(30)
       /       \
  Neerajpur   Sarasvatinagar
     /  \           \
 (15)/    \(40)      \(20)
   /       \           \
Sindhuvan   Varunabhoomi
     \           /
     (10)-------
In the water-scarce region of Jalvasthala, a central reservoir serves five key regions: Gangapur, Neerajpur, Sarasvatinagar, Sindhuvan, and Varunabhoomi. These regions differ in water demand and are connected via a network of roads forming a Directed Acyclic Graph (DAG):
Gangapur is the central reservoir and connects directly to Neerajpur (25 km) and Sarasvatinagar (30 km).
Neerajpur connects to Sindhuvan (15 km) and Varunabhoomi (40 km).
Sarasvatinagar connects directly to Sindhuvan (20 km) and Varunabhoomi (50 km).
Sindhuvan connects to Varunabhoomi (10 km), the farthest and highest-demand region.
The challenge is to minimize pipe line material required while ensuring sufficient water delivery to all regions,
prioritizing Varunabhoomi due to higher demand.
After laying the cables, If Sindhuvan goes for maintenace every tuesday, what should be the plan of supply for Varunabhoomi every tuesday?

"""

messages = [{"role": "user", "content": prompt}]
response = client.chat.completions.create(
    model="deepseek-reasoner",
    messages=messages
)

reasoning_content = response.choices[0].message.reasoning_content
content = response.choices[0].message.content

print('Think tokes that is reasoning logic: *****************************************')
print(reasoning_content)
print('final answer: *****************************************')
print(content)
```
